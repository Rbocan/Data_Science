# Modelos de Aprendizaje

## Métricas  a utilizar para evaluar los diferentes modelos de clasificación

* Antes de proceder con la configuración y entrenamiento de los modelos de predicción, en esta sección se describirán todas las métricas a utilizar para evaluar los diferentes modelos a utilizar en el proyecto. Esto con el objetivo de aclarar las métricas antes de utilizarlas.




### Accuracy Score (Puntaje de Exatitud)
* Es la fracción de predicciones de nuestro modelo que fueron clasificadas correctamente.
* Mientrás más se acerce este valor a 1, se tendrá una mejor exactitud.

$$
Accuracy = \frac{Número Predicciones Correctas}{Número Total Predicciones} \\
$$

Matemáticamente se puede ver de la siguiente forma:

$$ Accuracy(y,\hat{y}) = \frac{1}{n_{samples}} \sum_{i=0}^{n_{samples}} 1(\hat{y} = y)$$

* Donde 1(x) es la función de indicación, que nos permite contar si la predicción fue correcta o no.

### Balanced Accuracy Score
* Esta métrica nos permite lidiar con problemas de clasificación de conjuntos de datos desbalanceados.
* Evita estimaciones de rendimiento infladas en conjuntos de datos desequilibrados.
* Se puede definir como el la media aritmética de las sensibilidad (Precision) y especificidad (Recall) de recall obtenido en cada clase.
* El mejor valor es 1 y el pero 0.


$$ balancedAccuracy = \frac{1}{2} (\frac{TP}{TP+FN} + \frac{TN}{TN+FP} ) $$


### Kappa Score
* Se refiere al coeficiente kappa de Cohen. Esta es una medida estadística que mide el acuerdo entre anotadores en un problema de clasificación.
* Se define como: $$ k = \frac{(p_{o} - p_{e})}{(1- p_{e})}$$
* Donde $p_{o}$ es la probabilidad empírica de acuerdo en una etiqueta asignada a una muestra.
* Donde $p_{e}$ es el acuerdo esperado cuando ambos anotadores asignan la etiqueta aleatoriamente.
* Este coeficiente varía entre -1 y 1. Donde los coeficientes mayores a 0.8 son considerados como un buen acuerdo. Cero o menos significa que no hubo acuerd, prácticamente etiquetas aleatorias.

### Matthew Correlation Coeficient (MCC)
* El coeficiente de correlacion de Matthew se utiliza en aprendizaje automático como medición de la calidad de clasificaciones binarias o multiclase.
* Este toma en cueta los valores verdaderos y falsos positivos y negativos.
* Puede ser vista como una métrica balanceada la cual puede ser utilizada incluse si las clases son de diferentes tamaños.
* Este coeficiente varía entre -1 y +1. Donde +1 es una predicción perfecta, 0 es una predicción aleatoria y -1 es una predicción inversa.
* Se define como: $$ MCC = \frac{tp \times tn - fp \times fn}{\sqrt{(tp+fp)(tp+fn)(tn+fp)(tn+fn)}}$$


Extraído de: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.matthews_corrcoef.html#sklearn.metrics.matthews_corrcoef

### Precision Score (Puntaje de precisión)

* La precisión es la relación de $\frac{TP}{(TP + FP)}$, donde TP (True positive) son los verdaderos positivos y FP (False positive) son los falsos positivos.
* La precisión es la abilidad de clasificador de no etiquetar como positivo una muestra que es negativa.
* El mejor valor es de 1 y el peor 0.
* También se le conoce como Positive Predictive value (PPV)

Matemáticamente se puede ver de la siguiente forma:

$$\frac{TP}{(TP + FP)} = \frac{Number True Positives}{(Number True Positives + Number False Positives)} = \frac{Number True Positive}{Number Positive Calls} $$


**Nota:**
* Verdadero positivo >>  es cuando un se espera un 1 y se predice un 1
* Falso positivo >>  es cuando un se espera un 0 y se predice un 1
* Verdadero negativo >>  es cuando un se espera un 0 y se predice un 0
* Falso negativo >>  es cuando un se espera un 1 y se predice un 0


### Recall Score (Puntaje de Recuperación)

* El Recall es la relación de $\frac{TP}{(TP + FN)}$, donde TP (True positive) son los verdaderos positivos y FN (False negative) son los falsos negativos.
* El recall es la abilidad del clasificador para encontrar todos las muestras positivas.
* El mejor valor es de 1 y el peor 0.
* También se le conoce como True positive rate (TPV)

Matemáticamente se puede ver de la siguiente forma:

$$\frac{TP}{(TP + FN)} = \frac{Number True Positives}{(Number True Positives + Number False Negative)} = \frac{Number True Positive}{Number Expected Positive} $$


**Nota:**
* Verdadero positivo >>  es cuando un se espera un 1 y se predice un 1
* Falso positivo >>  es cuando un se espera un 0 y se predice un 1
* Verdadero negativo >>  es cuando un se espera un 0 y se predice un 0
* Falso negativo >>  es cuando un se espera un 1 y se predice un 0

### F1 Score (Puntuación F1)
* También se denomina puntuación F equilibrada o F medida F.
* Es la media armónica de la precisión y la recuperación.
* La puntuación F1 resulta útil cuando desea buscar un equilibrio entre la precisión y la recuperación.
* Una puntuación F1 alcanza el mejor valor en 1,00 y la peor puntuación en 0,00. Indica cuán preciso es el clasificador.

Matemática se puede obtener a partir de la siguiente fórmula:
$$F1 = 2 * \frac{(precision * recall)}{(precision + recall)}$$

### Matriz de Confusión
* La matriz de confusión evalúa la exactitud de la clasificación.
* Las filas (i) corresponden al valor esperado o real.
* Las columnas (j) corresponde al valor predicho.
* En una clasificación binaria, nos muestra la cuenta de Verdaderos negativos en C00, Falsos Positivos C10, Verdaderos Postivos C11 y Falsos Positivos C01.

## Decision Tree Classifier

```python
#Se importan todas las métricas a utilizar
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import balanced_accuracy_score, accuracy_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import matthews_corrcoef

#se importa el DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

Se iniciará con el algoritmo de Árbol de Decision (DecisionTreeClassifier).
* A continuación se iterará en el hiperparámetro max_depth del árbol de decisión creado.
* La idea es encontrar un valor óptimo que permitar obtener las mejores notas de las métricas utilizadas.
* Dado que conocemos que los datos están desbalanceado se utiliza el hiperparámetro class_weight= {0:1 , 1:10} como constante.

```python
#Se inicializan las listas de las métricas
training_score = []
test_score = []
balance_score = []
kappa_score = []
matthew_coef_score = []
precision = []
recall = []
f1 = []

#Se desea conocer el comportamiento al variar el max_depth de 1 a 35
depth_settings = range(1, 35)

for depth_tree in depth_settings:
    # se construye el modelo de clasificacion
    clf = DecisionTreeClassifier(max_depth=depth_tree, class_weight= {0:1 , 1:10}, random_state=10)
    clf.fit(X_train, y_train)
    y_predict = clf.predict(X_test)
    # se almacena el "training set accuracy"
    training_score.append(clf.score(X_train, y_train))
    # se almacena las diferentes métricas sobre los datos de testing
    test_score.append(clf.score(X_test, y_test))
    balance_score.append(balanced_accuracy_score(y_test, y_predict))
    kappa_score.append(cohen_kappa_score(y_test, y_predict))
    matthew_coef_score.append(matthews_corrcoef(y_test, y_predict))
    precision.append(precision_score(y_test, y_predict))
    recall.append(recall_score(y_test, y_predict))
    f1.append(f1_score(y_test, y_predict))

plt.plot(depth_settings, training_score, label="Training Accuracy")
plt.plot(depth_settings, test_score, label="Test Accuracy")
plt.plot(depth_settings, balance_score, label="Balanced Accuracy")
plt.plot(depth_settings, kappa_score, label="Kappa score")
plt.plot(depth_settings, matthew_coef_score, label="Matthew Coeficient")
plt.plot(depth_settings, recall, label="Recall Score")
plt.plot(depth_settings, precision, label="Precision Score")
plt.plot(depth_settings, f1, label="F1 Score")

plt.ylabel("Score")
plt.xlabel("max_depth")
plt.legend()
```

* **Importante:** De la gráfica anterior se puede observar que utilizar el Accuracy del Testing y del Training no es suficiente para determinar si el modelo está generalizando bien, esto se debe a que los datos están desbalanceados.
* A partir de una profundidad de 20, se empiezan a obtener todas las métricas mayores a 0.8.
* A una profundidad mayor de 30, no se observa un beneficio sobresaliente.
* Para continuar, se escogió utilizar un max_depth de 30.

```python
#Se inicializan las listas de las métricas
training_score = []
test_score = []
balance_score = []
kappa_score = []
matthew_coef_score = []
precision = []
recall = []
f1 = []


#Se varía el min_samples_leaf de 1 a 15
min_samples_leaf_settings = range(1, 15)

for min_leaf in min_samples_leaf_settings:
    # se construye el modelo de clasificacion
    clf = DecisionTreeClassifier(min_samples_leaf=min_leaf,max_depth=30, class_weight= {0:1 , 1:10}, random_state=10)
    clf.fit(X_train, y_train)
    y_predict = clf.predict(X_test)
    # se almacena el "training set accuracy"
    training_score.append(clf.score(X_train, y_train))
    # se almacena las diferentes métricas sobre los datos de testing
    test_score.append(clf.score(X_test, y_test))
    balance_score.append(balanced_accuracy_score(y_test, y_predict))
    kappa_score.append(cohen_kappa_score(y_test, y_predict))
    matthew_coef_score.append(matthews_corrcoef(y_test, y_predict))
    precision.append(precision_score(y_test, y_predict))
    recall.append(recall_score(y_test, y_predict))
    f1.append(f1_score(y_test, y_predict))

plt.plot(min_samples_leaf_settings, training_score, label="Training Accuracy")
plt.plot(min_samples_leaf_settings, test_score, label="Test Accuracy")
plt.plot(min_samples_leaf_settings, balance_score, label="Balanced Accuracy")
plt.plot(min_samples_leaf_settings, kappa_score, label="Kappa score")
plt.plot(min_samples_leaf_settings, matthew_coef_score, label="Matthew Coeficient")
plt.plot(min_samples_leaf_settings, recall, label="Recall Score")
plt.plot(min_samples_leaf_settings, precision, label="Precision Score")
plt.plot(min_samples_leaf_settings, f1, label="F1 Score")
plt.ylabel("score")
plt.xlabel("min_samples_leaf")
plt.legend()
```

* De la gráfica anterior se puede observar que a menor min_samples_leaf el modelo generaliza mejor.
* Para continuar, se escogió utilizar un min_samples_leaf de 1.

# Data balanceada utilizando un Submuestreo (Mayoría)

La idea de esta sección es implementar un submuestreo del set de datos original y analizar el comportamiento de todos los modelos utilizados anteriormente para tener una comparativa.

Esta es una estrategia simple y efectiva para problemas de clasificación desequilibrada.



Extraido de: https://machinelearningmastery.com/random-oversampling-and-undersampling-for-imbalanced-classification/


```python
import imblearn
print(imblearn.__version__)
```

```python
from imblearn.under_sampling import RandomUnderSampler
# Definiendo estrategia de Submuestreo
undersample = RandomUnderSampler(sampling_strategy='majority')
```

```python
# Definiendo estrategia de Submuestreo
undersample = RandomUnderSampler(sampling_strategy=0.5)
```

Si establecemos sampling_strategy en 0,5 en un conjunto de datos desequilibrados con 1000 ejemplos en la clase mayoritaria y 100 ejemplos en la clase minoritaria, habría 200 ejemplos para la clase mayoritaria en el conjunto de datos transformado ( 100/200 = 0,5 ).

#Conclusiones

* La métrica de accuracy no es suficiente para poder concluir que un modelo generaliza bien cuando se tienen datos desbalanceados. Usualmente se obtendrán valores bastante altos esto debido a todos los aciertos que pueda aportar la clase mayoritaria.

* La métrica de Balanced Accuracy y los coeficientes Kappa y MCM nos ayudarón a poder identificar qué tan bien se estaba generalizado debido a que la exactitud de testing y training eran altos a causa del desbalanceo.

* Los algoritmos de Decision Tree y Random Forest son los que mejor generalizan y se adjustan a los datos utilizados.

* La técnica de iterar sobre los diferentes hiper-parámetros durante la configuración de los modelos, nos permitió encontrar los valores que aportaban a una mejor generalización.

* En el caso de los Decision Tree y Random Forest, el hiper-parámetro "class_weight" nos permitió poder indicar que los datos estaban desbalanceado es una proporción de 1:10 {0:1 , 1:10}. Esto sin lugar a duda nos ayudó a lidiar con el desbalanceo con la configuración de estos modelos.

* A pesar de que KNN neighbors consigue valores regulares de los coeficientes, al tener el hiperparámetro n_neighbors=1 como el que generaliza mejor, se puede pensar que cada predicción buscará el dato más parecido para poder clasificar.

* Los métodos de Naive Bayes utilizados (Gausiano, Complement, Bernoulli) no son los mejores para los datos utilizados, no se logró obtener buenos coeficientes de Kappa y de MCM ni un una buena exactitud balanceada.

* Los algoritmos de kernel SVM tampoco se acoplaron bien a los datos. Al utilizar los datos completos, los coefientes de Kappa y MCM se mantuvieron en cero.

* A pesar de que se obtuvieron buenos puntajes de las métricas con los datos desbalanceado, se decidió realizar la prueba de balanceo llamada Undersampling. Sin embargo, de igual manera se pudo haber utilizado otra técnias como Oversampling de clase minoritaria. Optamos por el undersampling ya que teníamos una gran cantidad de datos.
